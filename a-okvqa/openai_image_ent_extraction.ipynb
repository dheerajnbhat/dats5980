```python
AZURE_OPENAI_API_VERSION = "2025-02-01-preview"
AZURE_OPENAI_ENDPOINT = ""
AZURE_OPENAI_DEPLOYMENT_NAME = "gpt-4.1"
AZURE_OPENAI_API_KEY = ""
```

```python
llm = AzureOpenAI(
    api_version=AZURE_OPENAI_API_VERSION,
    azure_endpoint=AZURE_OPENAI_ENDPOINT,
    azure_deployment=AZURE_OPENAI_DEPLOYMENT_NAME,
    api_key=AZURE_OPENAI_API_KEY
)
```

```python
completion = llm.chat.completions.create(
    model=AZURE_OPENAI_DEPLOYMENT_NAME,
    messages=[
        {
            "role": "user",
            "content": "who won the cricket world cup in 1983?",
        },
    ],
    temperature=0
)
completion.choices[0].message.content
```

```python
# image_path = r"C:\workspace\misc\5980\sample_image.jpg"
image_path = r"C:\workspace\misc\5980\coco\val2017\000000461751.jpg"

# Read and base64-encode the image
with open(image_path, "rb") as f:
    image_base64 = base64.b64encode(f.read()).decode("utf-8")
```

```python
response = llm.chat.completions.create(
    model=AZURE_OPENAI_DEPLOYMENT_NAME,
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "What is in the motorcyclist's mouth"},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{image_base64}"
                    },
                },
            ],
        }
    ],
    temperature=0
)
response.choices[0].message.content
```
