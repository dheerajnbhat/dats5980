```python
dataset_path = r"data/val_predictions_mcq.pkl"
dataset_json = r"data/val_predictions_mcq.json"

with open(dataset_path, "rb") as fp:
    data = pickle.load(fp)

final_data = {
    i[0]['question_id']: {
        "multiple_choice": i[1].split("Assistant:")[1].strip()
    } for i in data
}
print(len(final_data))

with open(dataset_json, "w") as fp:
    json.dump(final_data, fp)
```
